```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect('travel.db')
```

```python
name = ["Liza", "Matt", "Pit", "Kurt", "Lia"]
surname = ["Simpson", "Walker", "Berg", "Walter", "Mishelle"]
passenger_df = pd.DataFrame({'surname':surname ,'name': name})
passenger_df.to_sql('passenger', con, index_label="id", if_exists='replace') 
pd.read_sql("SELECT * FROM passenger", con) 
```

```python
company = ["Pobeda", "UTair", "S7 Airline"]
company_df = pd.DataFrame({'name': company})
company_df.to_sql('company', con, index_label="id", if_exists='replace') 
pd.read_sql("SELECT * FROM company", con) 
```

```python
from random import randrange
from datetime import timedelta
from datetime import datetime

def random_date(start, end):
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

```python
time_in = pd.DataFrame()
time_in['time_in'] = None
for i in range(10):
    d1 = datetime.strptime('1/1/2008 2:50 PM', '%m/%d/%Y %I:%M %p')
    d2 = datetime.strptime('1/1/2008 7:50 PM', '%m/%d/%Y %I:%M %p')
    a=random_date(d1, d2)
    b={'time_in':a}
    time_in=time_in.append(b, ignore_index=True)

time_out = pd.DataFrame()
time_out['time_out'] = None
for i in range(10):
    d1 = datetime.strptime('1/1/2008 1:30 PM', '%m/%d/%Y %I:%M %p')
    d2 = datetime.strptime('1/1/2008 2:50 PM', '%m/%d/%Y %I:%M %p')
    a=random_date(d1, d2)
    b={'time_out':a}
    time_out=time_out.append(b, ignore_index=True)
```

```python
companyy = ["0", "1", "2","2","1","0","2","1","0","2"]
plane = ["Boeing 737-800", "Mi-26T", "Airbus A319", "Embtaer 170", "Mi-171", "Boeing 737-800", "Airbus A319", "Boeing 737-400", "Boeing 737-800","Airbus A319"]
town_from = ["SPB","Paris","Msk","Paris","SPB","Yar","London","Amsterda","Novosib","SPB"]
town_to = ["Msk", "Msk", "Paris","Msk","London","SPB","Msk","Msk","SPB","Amsterdam"]
trip_df = pd.DataFrame({'company': companyy, 'plane':plane, 'town_from':town_from,'town_to':town_to })
trip_df=pd.concat([trip_df,time_out,time_in],axis=1)
trip_df.to_sql('trip', con, index_label="id", if_exists='replace') 
pd.read_sql("SELECT * FROM trip", con) 
```

```python
from random import randint
import random

places = ['1a','1b','1c','2a','2b','2c','3a','3b','3c']

place_df = pd.DataFrame()
place_df['place'] = None

trip_df = pd.DataFrame()
trip_df['trip'] = None

passeng_df = pd.DataFrame()
passeng_df['passenger'] = None


for i in range(10):
    a = randint(0,9)
    b = {'trip':a}
    trip_df=trip_df.append(b, ignore_index=True)
    c=randint(0,4)
    d={'passenger':c}
    passeng_df=passeng_df.append(d, ignore_index=True)
    e=random.choice(places)
    f={'place':e}
    place_df=place_df.append(f, ignore_index=True)
    

pass_in_trip_df = pd.DataFrame()
pass_in_trip_df=pd.concat([trip_df,passeng_df,place_df],axis=1)
pass_in_trip_df
pass_in_trip_df.to_sql('pass_in_trip', con, index_label="id", if_exists='replace') 
pd.read_sql("SELECT * FROM pass_in_trip", con) 
```

```python
pd.read_sql("SELECT DISTINCT name FROM company JOIN trip ON company.id = trip.company WHERE plane LIKE '%Boeing%';", con)
```
